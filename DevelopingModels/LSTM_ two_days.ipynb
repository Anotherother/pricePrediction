```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
from scipy import stats
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential, load_model, save_model
import datetime 
import seaborn as sns
from pylab import rcParams
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
import time

import plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot

import sqlalchemy

init_notebook_mode(connected=True)

%matplotlib inline

sns.set(style='whitegrid', palette='muted', font_scale=1.5)

rcParams['figure.figsize'] = 14, 8

RANDOM_SEED = 42
NUMDAY_PREDICTION = 2
WINDOW = 22
```

```python

def connect(user, password, db, host: str, port: int, echo=False):
    url = 'postgresql+psycopg2://{}:{}@{}:{}/{}'
    url = url.format(user, password, host, port, db)
    eng = sqlalchemy.create_engine(url, client_encoding='utf8', echo=echo)
    meta = sqlalchemy.MetaData(bind=eng)
    return eng, meta

def get_data_frame(pair: str = 'USDT_BTC', exchange: str = 'poloniex') -> pd.DataFrame:

    engine, meta = connect(user='postgres', password='password', db='btccandles', host='94.230.125.199', port=16432)
    df = pd.read_sql_query(
        "SELECT date, time, open, close, low, high, volume, pair.\"name\""
        "FROM candlestick, pair WHERE candlestick.pair_id = pair.id AND pair.id IN ("
        "SELECT pair.id FROM pair, exchange WHERE ("
        "SELECT pair.id FROM pair WHERE pair.name = '" + pair +
        "') = pair.alias_id AND pair.exchange_id = (SELECT exchange.id FROM exchange WHERE exchange.\"name\" = '" + exchange + "')) ORDER BY candlestick.date;",
        con=engine)
    return df

def LoadData(data, train_size, n_day = NUMDAY_PREDICTION, TestTrain = False):
    
    amount_of_features = data.shape[1] 
    data_mat = data.as_matrix() 

    frames = []
    for index in range(len(data_mat) - WINDOW):
        frames.append(data_mat[index: index + WINDOW ])

    frames = np.array(frames)
    if TestTrain == False:
        x  = frames[:-NUMDAY_PREDICTION,:,:-1] # delete feature, which we want predict and add it to 
                                   # labels - y_train
        
        temp_y  = frames[:, -1][:,-1]   # lat index - index with value, 
                                  # which we want predict. 4 - index for 'close'

        y = []
        for index in range(len(temp_y) - NUMDAY_PREDICTION):
            y.append(temp_y[index: index + NUMDAY_PREDICTION ])
        y = np.array(y)

        return x,y
    
    train_split = int(round(train_size * data.shape[0]))
        
    if TestTrain == True:    
        x_train = frames[:train_split,:,:-1]
        x_test = frames[train_split:-NUMDAY_PREDICTION,:,:-1]
        
        
        
        temp_y  = frames[:, -1][:,-1]   # lat index - index with value, 
                                  # which we want predict. 4 - index for 'close'

        y = []
        for index in range(len(temp_y) - NUMDAY_PREDICTION):
            y.append(temp_y[index: index + NUMDAY_PREDICTION ])
        y = np.array(y)

        y_train = y[:train_split,:n_day,4]
        y_test = y[train_split:-NUMDAY_PREDICTION,:n_day,4]

        print (x_train.shape)
        print (x_test.shape)
        return x_train,x_test, y_train, y_test
"""
Собираем LSTM
"""

def build_model(input_shape):
    d = 0.2
    model = Sequential()
    
    model.add(LSTM(128, input_shape=input_shape, return_sequences=True))
    model.add(Dropout(d))
        
    model.add(LSTM(128, input_shape=input_shape, return_sequences=False))
    model.add(Dropout(d))
        
    model.add(Dense(32,kernel_initializer="normal",activation='relu'))        
    model.add(Dense(NUMDAY_PREDICTION,kernel_initializer="normal",activation='linear'))
    
    model.compile(loss='mse',optimizer='adam', metrics=['accuracy'])
    
    return model

def predictDays(pathModel, data, scaler):
    m = load_model(pathModel)    
    prediction = []
    lastbatch = np.array(data[['open', 'low', 'high', 'volume']][-WINDOW:])
    
    prediction = m.predict([lastbatch.reshape(1,22, 4)])
    
    return np.array(prediction)
```

```python
"""
typeBlockchain:
USDT_BTC
USDT_LTC
USDT_ETH
USDT_ETC
USDT_XRP
"""
typeBlockchain = 'USDT_BTC'
stock = 'poloniex'

# read data from database
df = get_data_frame(typeBlockchain, stock)

#initialise scalers for data normalisation
x_scaler = MinMaxScaler()
y_scaler = MinMaxScaler()

data = df[['open', 'low', 'high', 'volume']].copy()
y = df['close'].copy()

data[['open', 'low', 'high', 'volume']] = x_scaler.fit_transform(data)

y = y_scaler.fit_transform(y)
data['close'] = y

X_train, y_train = LoadData(data, train_size = 0.5, n_day = NUMDAY_PREDICTION, TestTrain = False)

model = build_model(input_shape=(WINDOW, 4)) # feature number here
```

```python
print('START FIT MODEL...')

start = time.time()
model.fit(X_train, y_train, batch_size=32, epochs=500,
          verbose=0)
end = time.time()

print ('Learning time: ', end-start)
```

```python
today = time.strftime("_%d_%m_%Y")

pathModel = "../models/model_5f_" + typeBlockchain + today +".h5"
save_model(model, pathModel)

#model = load_model(pathModel)

trainPredict = model.predict(X_train)
trainPredict = y_scaler.inverse_transform(trainPredict)
trainY = y_scaler.inverse_transform(y_train)
```

```python
trainScore = metrics.mean_squared_error(trainY[:,0], trainPredict[:,0]) ** .5
print('Train Score: %.2f RMSE' % (trainScore))

prices = df.close.values.astype('float32')
prices = prices.reshape(len(prices), 1)

trainPredictPlot = np.empty_like(prices)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[WINDOW:len(trainPredict)+WINDOW, :] = trainPredict[:,0].reshape(-1,1)


lastDate =str(df.date[df.last_valid_index()]).split('-')

predictNday =  (predictDays(pathModel, data, y_scaler))
```
